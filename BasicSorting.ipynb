### 그래프 그리기 코드

matplotlib 을 이용하여 그래프를 그리기 위한 코드입니다. <br>
배열 A와 제목을 입력받아 바 그래프를 그려줍니다.

````python
import matplotlib.pyplot as plt

def drawPlot(A, title="plot"):
    x = [x for x in range(len(A))]

    plt.bar(x, A, label="Blue Bar", color='b')
    plt.xlabel("index")
    plt.ylabel("value")
    plt.title(title)
    plt.show()
````
