### 그래프 그리기 코드

matplotlib 을 이용하여 그래프를 그리기 위한 코드입니다. <br>
배열 A와 제목을 입력받아 바 그래프를 그려줍니다.

```python
import matplotlib.pyplot as plt

def drawPlot(A, title="plot"):
    x = [x for x in range(len(A))]

    plt.bar(x, A, label="Blue Bar", color='b')
    plt.xlabel("index")
    plt.ylabel("value")
    plt.title(title)
    plt.show()
```

### 선택정렬 코드

![](Images/selectionsort.png)

위의 알고리즘을 참고하여 코드를 작성하세요

```python
def selectionSort(A):
	for last in range(len(A)-1, 0, -1):
		k = theLargest(A, last)	# A[0...last] 중 가장 큰 수 A[k] 찾기
		A[k], A[last] =  A[last], A[k]

def theLargest(A, last:int) -> int:	# A[0...last]에서 가장 큰 수의 인덱스를 리턴한다
	largest = 0
	for i in range(last+1):
		if A[i] > A[largest]:
			largest = i
	return largest
```

### 테스트하기
작성한 코드를 이용하여 정렬을 수행합니다.

```python

print("selectionSort test")
A = [35, 24, 16, 21, 4, 72, 23, 9, 23, 14, 58, 12, 1]
print("A[]:	   ", A)

drawPlot(A, "Before")

selectionSort(A)
print("Sorted A[]:", A)

drawPlot(A, "After")
```
